```python
import os
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()

# âœ… Securely load your key from environment variables
api_key = os.getenv("NVIDIA_API_KEY")
if not api_key:
    raise ValueError("Please set NVIDIA_API_KEY in your environment.")

client = OpenAI(
    base_url="https://integrate.api.nvidia.com/v1",
    api_key=api_key
)

completion = client.chat.completions.create(
    model="nvidia/llama-3.1-nemotron-70b-instruct",  # or your chosen model
    messages=[{"role": "user", "content": "Hello, world!"}],
    temperature=0.5,
    top_p=0.9,
    max_tokens=512,
    stream=True
)

# Stream and print response chunks as they arrive
for chunk in completion:
    delta = chunk.choices[0].delta
    if getattr(delta, "content", None):
        print(delta.content, end="")
```

```python
from langchain_community.document_loaders import TextLoader

# Try UTF-8 first
raw_documents = TextLoader("tagged_description.txt", encoding="utf-8").load()
```

```python
from langchain.text_splitter import CharacterTextSplitter

# âœ… Set a realistic chunk size (e.g., 500 characters)
text_splitter = CharacterTextSplitter(chunk_size=300,chunk_overlap=50,separator="\n")
```

```python

# Assuming raw_documents is a list of Document or string objects
documents = text_splitter.split_documents(raw_documents)

# Optional: Inspect the results
print(f"Number of documents: {len(documents)}")
print(documents[0])
```

```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import Chroma

# Load a sentence-transformers model (you can change this to another HF-compatible model)
embedding = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

# Now use it with Chroma
db_books = Chroma.from_documents(
    documents,
    embedding=embedding
)
```

```python
query = " A book to teach children about nature "
docs = db_books.similarity_search(query , k = 10)
docs
```

```python
books[books["isbn13"] == int(docs[0].page_content.split()[0].strip())]
```

```python
def retrieve_semantic_recommendations(
    query:str,
    top_k: int = 10 ,
) ->pd.DataFrame:
    recs = db_books.similarity_search(query , k = 50)
    books_list = []
    for i in range(0 , len(recs)):
        books_list+= [int(recs[i].page_content.strip('"').split()[0])]
    return books[books["isbn13"].isin(books_list)].head(top_k)
```

```python
retrieve_semantic_recommendations("A book to teach children about nature")
```
